# Pandas — Vectorization vs Loops

```python
# Initialize

#Imports
import pandas as pd
pd.set_option("display.max_columns", None)
# Load Data
employees = pd.read_csv("../data/employees.csv")
departments = pd.read_csv("../data/departments.csv")
sales = pd.read_csv("../data/sales.csv")
```

```python
# WRONG APPROACH — Python loop

bonus_category = []

for salary in employees["salary"]:
    if salary > 80000:
        bonus_category.append("High")
    else:
        bonus_category.append("Normal")

employees["bonus_category_loop"] = bonus_category

employees.head()
```

```python
# CORRECT APPROACH — Vectorized

employees["bonus_category_vectorized"] = (
    employees["salary"] > 80000
)

employees["bonus_category_vectorized"] = (
    employees["bonus_category_vectorized"]
        .map({True: "High", False: "Normal"})
)

employees.head()
```

```python
# Using Numpy

import numpy as np

employees["bonus_category_np"] = np.where(
    employees["salary"] > 80000,
    "High",
    "Normal"
)
```

```python
# .apply() -- use it properly

# wrong approach

employees["category_apply"] = employees["salary"].apply(
    lambda x: "High" if x > 80000 else "Normal"
)
'''
This is cleaner than loop.

But:

⚠ Still Python-level execution.
'''
```

```python
## Date Logic

# Calculate employee tenure in days

current_date = pd.Timestamp("2026-02-22")

employees["joining_date"] = pd.to_datetime(
    employees["joining_date"],
    errors="coerce"
)

employees["tenure_days"] = (
    current_date - employees["joining_date"]
).dt.days

employees.head()

# Python loop cateogrization

tenure_bucket = []

for days in employees["tenure_days"]:
    if days > 1000:
        tenure_bucket.append("Senior")
    else:
        tenure_bucket.append("Junior")
```

```python
## Vectorized

employees["tenure_bucket"] = np.where(
    employees["tenure_days"] > 1000,
    "Senior",
    "Junior"
)
```

```python
## Aggregation without loops

# Bad approach

result = {}
for dept in employees["department"].unique():
    result[dept] = employees[employees["department"] == dept]["salary"].mean()

# Correct Approach
employees.groupby("department")["salary"].mean()
```
