# Pandas — Missing Values & Data Type Handling

```python
# Initialize

#Imports
import pandas as pd
pd.set_option("display.max_columns", None)
# Load Data
employees = pd.read_csv("../data/employees.csv")
departments = pd.read_csv("../data/departments.csv")
sales = pd.read_csv("../data/sales.csv")
```

```python
# Simulating missing data

employees.loc[3, "salary"] = None
employees.loc[5, "department"] = None

employees.head(6)
```

```python
# Count missing values per column

employees.isnull().sum()
```

```python
# Percentage of missing values

(employees.isnull().mean() * 100).round(2)
```

```python
#Option 1: Drop rows (dangerous)

employees_cleaned = employees.dropna()
employees_cleaned.shape
```

```python
# Option 2 - Fill with business logic - Fill salary with median (safer than mean)

employees["salary"] = employees["salary"].fillna(
    employees["salary"].median()
)
employees
```

```python
# Option 3 - Fill categorical
employees["department"] = employees["department"].fillna("Unknown")
employees
```

## Data Type Conversion


```python
employees.dtypes
```

```python
# Convert to datetime

employees["joining_date"] = pd.to_datetime(
    employees["joining_date"],
    errors="coerce"
)
```

```python
# Convert salary safely

employees["salary"] = pd.to_numeric(
    employees["salary"],
    errors="coerce"
)
```

```python
#Type Validation Before Join
sales["emp_id"] = sales["emp_id"].astype(int)
employees["emp_id"] = employees["emp_id"].astype(int)
```
