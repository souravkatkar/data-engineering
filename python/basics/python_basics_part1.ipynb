## Data Types

Python provides multiple data types to store and process data.
In data engineering, choosing the right data type impacts:
- Readability
- Performance
- Correctness


```python
## Data Types Examples


# Integer
a = 10
print(a, type(a))

# Float
b = 10.5
print(b, type(b))

# String
c = "hello"
print(c, type(c))

# Boolean
d = True
print(d, type(d))

# NoneType
e = None
print(e, type(e))
```

## Collections Overview

```python
## Basic Creation

# List (ordered, mutable)
my_list = [1, 2, 3, 3]
print(my_list, type(my_list))

# Tuple (ordered, immutable)
my_tuple = (1, 2, 3, 3)
print(my_tuple, type(my_tuple))

# Set (unordered, unique elements)
my_set = {1, 2, 3, 3}
print(my_set, type(my_set))

# Dictionary (key-value pairs)
my_dict = {"a": 1, "b": 2}
print(my_dict, type(my_dict))
```

```python
## Mutability Test

# List is mutable
my_list.append(4)
print("List:", my_list)

# Tuple is immutable (this will fail if uncommented)
# my_tuple.append(4)

# Set is mutable
my_set.add(4)
print("Set:", my_set)

# Dictionary is mutable
my_dict["c"] = 3
print("Dict:", my_dict)
```

## Indexing and Slicing

```python
## Indexing Basics

data = [10, 20, 30, 40, 50]

print(data[0])   # first element
print(data[2])   # third element
print(data[-1])  # last element
print(data[-2])  # second last element
```

```python
## Slicing Basics

print(data[1:4])   # index 1 to 3
print(data[:3])    # start to index 2
print(data[2:])    # index 2 to end
print(data[:])     # full copy
```

```python
## Step & Reverse

print(data[::2])    # every 2nd element
print(data[::-1])  # reverse list
```

## Conditional Statements

```python
## Basic if / else

x = 10

if x > 5:
    print("x is greater than 5")
else:
    print("x is less than or equal to 5")
```

```python
## if / elif / else

score = 75

if score >= 90:
    print("Grade A")
elif score >= 75:
    print("Grade B")
elif score >= 60:
    print("Grade C")
else:
    print("Fail")
```

```python
## Comparison & Logical Operators

age = 25
has_id = True

if age >= 18 and has_id:
    print("Allowed entry")
else:
    print("Not allowed")
```

```python
## Truthy & Falsy

values = [0, 1, "", "hello", [], [1, 2], None]

for v in values:
    if v:
        print(v, "is Truthy")
    else:
        print(v, "is Falsy")
```

## Loops

```python
## for loop

numbers = [10, 20, 30, 40]

for n in numbers:
    print(n)
```

```python
## for loop with range

for i in range(5):
    print(i)
```

```python
## Loop with index (enumerate)

fruits = ["apple", "banana", "mango"]

for index, fruit in enumerate(fruits):
    print(index, fruit)
```

```python
## while loop

count = 0

while count < 3:
    print(count)
    count += 1
```

```python
## break & continue

for i in range(5):
    if i == 3:
        break
    print(i)

for i in range(5):
    if i == 3:
        continue
    print(i)
```

## Functions

```python
## Basic Function

def greet():
    print("Hello, World!")

greet()
```

```python
## Function with Parameters

def greet_user(name):
    print(f"Hello, {name}")

greet_user("Alice")
greet_user("Bob")
```

```python
## return vs print

def add(a, b):
    return a + b

result = add(3, 4)
print(result)
```

```python
## return vs print

def add_print(a, b):
    print(a + b)

x = add_print(3, 4)
print(x)
```

```python
## Default Arguments

def greet_country(name, country="India"):
    print(f"{name} is from {country}")

greet_country("Rahul")
greet_country("John", "USA")
```

```python
## Multiple Returns

def calculate(a, b):
    sum_val = a + b
    diff_val = a - b
    return sum_val, diff_val

s, d = calculate(10, 5)
print(s, d)
```

```python
## *args (Variable Positional Arguments)

def add_numbers(*args):
    print(args)
    total = 0
    for num in args:
        total += num
    return total

print(add_numbers(1, 2))
print(add_numbers(1, 2, 3, 4))
```

```python
## **kwargs (Variable Keyword Arguments)

def print_details(**kwargs):
    print(kwargs)
    for key, value in kwargs.items():
        print(key, ":", value)

print_details(name="Alice", age=25, city="Pune")


```

```python
## Mixing Normal Args, *args, **kwargs

def demo(a, b, *args, **kwargs):
    print("a:", a)
    print("b:", b)
    print("args:", args)
    print("kwargs:", kwargs)

demo(1, 2, 3, 4, x=10, y=20)
```

## Comprehensions (List / Dict / Set)

```python
## List Comprehension

numbers = [1, 2, 3, 4, 5]

squares = [n * n for n in numbers]
print(squares)
```

```python
## List Comprehension with Condition

even_numbers = [n for n in numbers if n % 2 == 0]
print(even_numbers)
```

```python
## Dict Comprehension

names = ["alice", "bob", "charlie"]

name_lengths = {name: len(name) for name in names}
print(name_lengths)
```

```python
## Set Comprehension

nums = [1, 2, 2, 3, 3, 4]

unique_squares = {n * n for n in nums}
print(unique_squares)
```
